# EDA with Visuals
Create visualizations to answer the questions below. Use the `winequality_edited.csv` dataset, which you should've created  in the previous quiz.

```python
import pandas as pd 
df=pd.read_csv('data/winequality_edited.csv')
df.head()
```

### Which of the following feature variables appear skewed to the right? Fixed Acidity, Total Sulfur Dioxide, pH, Alcohol _____________________ (answer using plots)

```python
import matplotlib.pyplot as plt

# create histograms of the feature variables
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axes = axes.flatten()
features = ['fixed_acidity', 'total_sulfur_dioxide', 'pH', 'alcohol']

for i, feature in enumerate(features):
    axes[i].hist(df[feature], bins=30, color='blue', alpha=0.5)
    axes[i].set_xlabel(feature, fontsize=12)
    axes[i].set_ylabel('Frequency', fontsize=12)

plt.tight_layout()
plt.show()

# total_sulfur_dioxide, alcohol
```

### Which of the following is most likely to have a positive impact on quality? Volatile Acidity, Residual Sugar, pH, Alcohol _____________________________ (answer using plots)

```python
# create scatter plots of feature variables against quality rating
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
axes = axes.flatten()
features = ['volatile_acidity', 'residual_sugar', 'pH', 'alcohol']

for i, feature in enumerate(features):
    axes[i].scatter(df[feature], df['quality'], color='blue', alpha=0.5)
    axes[i].set_xlabel(feature, fontsize=12)
    axes[i].set_ylabel('Quality', fontsize=12)

plt.tight_layout()
plt.show()
```
